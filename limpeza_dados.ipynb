```python
import pandas as pd
```

```python
nutricao_com_trad=pd.read_csv('dados_nutricionais_comunidades_tradicionais.csv',low_memory=False)
```

### PORCENTAGEM DE IMC POR ANO POR COMUNIDADE

```python
# PEGA SOMENTE CASOS DE CRIANÇAS MENORES DE 12 ANOS
nutricao_com_trad=nutricao_com_trad[nutricao_com_trad['idade'] <= 12]
```

```python
# OBSERVAÇÃO: CADA INDIVÍDUO TEM UMA ID. POR ISSO, PEGAR A PORCENTADEM DE CADA FAIXA POR ANO PODE MOSTRAR UM VALOR ENVIESADO
## (O MESMO INDIVÍDUO PODE SER LIDO DUAS OU MAIS VEZES EM UM ANO)

# CONTA A QUANTIDADE DE INDIVÍDUOS POR ANO POR MÊS

conta_individuos_ano_mes=nutricao_com_trad.groupby(['ano', 'mes'])['id_individuo'].nunique().unstack(fill_value=0)
conta_individuos_ano_mes=conta_individuos_ano_mes.reset_index()
conta_individuos_ano_mes
```

```python
# MOSTRA QUANTIDADE DE INDIVÍDUOS COM MEDIÇÃO REPETIDA EM CADA ANO

anos=[x for x in range(2008,2024)]

for ano in anos:
    df=nutricao_com_trad[nutricao_com_trad['ano'] == ano]
    print(f"Quantidade máxima de indivíduos repetidos no ano {ano}: {df['id_individuo'].value_counts().max()}")
    print('-----------------------------------')
```

```python
# DERRUBA DUPLICATAS EM CADA ANO E DEIXA SOMENTE O ÚLTIMO VALOR QUE APARECE

# ORDENA POR ANO/MES
organiza_duplicata = nutricao_com_trad.sort_values(by=['ano', 'mes'])

# REMOVE DUPLICATAS DE INDIVÍDUOS POR ANO, MANTENHO O ÚLTIMO REGISTRO
derruba_duplicata = organiza_duplicata.drop_duplicates(subset=['ano', 'id_individuo'], keep='last')

derruba_duplicata
```

```python
# QUANTIDADE DE POPULAÇÃO POR ANO POR COMUNIDADE

quantidade_comunidade=derruba_duplicata.groupby(['povo_comunidade', 'ano']).size().unstack(fill_value=0)
quantidade_comunidade=quantidade_comunidade.reset_index()
quantidade_comunidade
```

```python
# CONTA A QUANTIDADE DE TIPOS DE PESO POR ANO POR COMUNIDADE 

conta_comunidade_ano = derruba_duplicata.groupby(['povo_comunidade', 'estado_nutricional_peso_idade_crianca', 'ano']).size().unstack(fill_value=0)
conta_comunidade_ano=conta_comunidade_ano.reset_index()
conta_comunidade_ano
```

```python
# MODIFICA PARA ESTRUTURA DO FLOURISH

df_melted = pd.melt(conta_comunidade_ano, id_vars=['povo_comunidade', 'estado_nutricional_peso_idade_crianca'], 
                    var_name='ano', value_name='quantidade')

# Pivotando a tabela para ter uma coluna para cada estado nutricional
df_pivot = df_melted.pivot_table(index=['ano', 'povo_comunidade'], 
                                 columns='estado_nutricional_peso_idade_crianca', 
                                 values='quantidade', 
                                 aggfunc='sum',fill_value=0).reset_index()
```

```python
# APLICA NOME DAS COMUNIDADES

dicionario={
    1: "POVOS QUILOMBOLAS",
    2: "AGROEXTRATIVISTAS",
    3: "CAATINGUEIROS",
    4: "CAIÇARAS",
    5: "COMUNIDADES DE FUNDO E FECHO DE PASTO",
    6: "COMUNIDADES DO CERRADO",
    7: "EXTRATIVISTAS",
    8: "FAXINALENSES",
    9: "GERAIZEIROS",
    10: "MARISQUEIROS",
    11: "PANTANEIROS",
    12: "PESCADORES ARTESANAIS",
    13: "POMERANOS",
    14: "POVOS CIGANOS",
    15: "POVOS DE TERREIRO",
    16: "QUEBRADEIRAS DE COCO-DE-BABAÇU",
    17: "RETIREIROS",
    18: "RIBEIRINHOS",
    19: "SERINGUEIROS",
    20: "VAZANTEIROS",
    21: "OUTROS"
    }


df_pivot['povo_comunidade']=df_pivot['povo_comunidade'].replace(dicionario)
df_pivot.to_csv('dados_comunidade_ano.csv',index=False)
```

```python
import matplotlib.pyplot as plt


# Supondo que estado_nutricional_comunidade já esteja definido
estado_nutricional_comunidade_filtra = estado_nutricional_comunidade[['ano', 'povo_comunidade', 'Baixo peso para a idade %',
                                                                      'Muito baixo peso para a idade %',
                                                                      'Peso adequado para idade %',
                                                                      'Peso elevado para a idade %']]

estado_nutricional_comunidade_flourish = pd.pivot_table(estado_nutricional_comunidade_filtra, 
                                                        index='ano', 
                                                        columns='povo_comunidade', 
                                                        values=['Baixo peso para a idade %',
                                                                'Muito baixo peso para a idade %',
                                                                'Peso adequado para idade %',
                                                                'Peso elevado para a idade %'])
estado_nutricional_comunidade_flourish = estado_nutricional_comunidade_flourish.reset_index()


# Reformatando os dados para o gráfico de área
estados = ['Baixo peso para a idade %', 
           'Muito baixo peso para a idade %', 
           'Peso adequado para idade %', 
           'Peso elevado para a idade %']

# Iterando por cada comunidade
for comunidade in estado_nutricional_comunidade_filtra['povo_comunidade'].unique():
    df_comunidade = estado_nutricional_comunidade_flourish.xs(comunidade, level=1, axis=1)
    
    # Plotando o gráfico de área
    plt.figure()
    plt.stackplot(df_comunidade['ano'], 
                  df_comunidade[estados].T, 
                  labels=estados, 
                  alpha=0.8)
    plt.title(f'Gráfico de Área de Proporção - {comunidade}')
    plt.xlabel('Ano')
    plt.ylabel('Proporção (%)')
    plt.legend(loc='upper left')
    plt.show()
```
